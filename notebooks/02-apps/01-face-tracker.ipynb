```python
from time import sleep

from hailo_apps.apps import FaceTracker
from hailo_apps.servos import ServoAngles
from hailo_apps.meta.interfaces import RotatorParams, ImageSize

from rich.pretty import pprint
```

```python
face_tracker = FaceTracker(
    init_servo_angles=ServoAngles(),
    rotator_params=RotatorParams(),
    image_size=ImageSize(
        width=1024,
        height=1024,
    ),
    debug_mode=True,
    history_length=300,  # 10 seconds at 30 fps.
)
```

```python
face_tracker.run()
# sleep(20)
```

```python
face_tracker.stop()
```
